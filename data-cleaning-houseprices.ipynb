#### Hora de limpar e transformar

```python
# Para deixar tugo igual, primeiro vou incluir os campos que faltam no registro de endereço
```

```python
# cria uma nova coluna de endereço

df['endereco2'] = df['endereco']
```

```python
# substitui ',' por '-'
# 
df['endereco2'] = df['endereco'].str.replace(',','-')
df.loc[df['endereco2'].str.count('-') == 1, 'endereco2'] = ' - - - ' + df['endereco2'].astype(str)
df.loc[df['endereco2'].str.count('-') == 2, 'endereco2'] = ' - - ' + df['endereco2'].astype(str)
df.loc[df['endereco2'].str.count('-') == 3, 'endereco2'] = ' - ' + df['endereco2'].astype(str)
df['endereco2']
```

```python
# novo DataFrame com os campos do endereço separados

novo = df['endereco2'].str.split("-", n = 4,  expand = True)
colunas = ['rua','numero','bairro','cidade','uf']
novo.columns = colunas
```

```python
novo
```

```python
# limpar e organizar os campos de endereço

novo.loc[novo['rua'] == '', 'rua'] = novo['numero']
novo['numero'] = novo['numero'].str.extract('(\d+)',expand=False)
novo
```

```python
# apagar a culuna criada para trabalhar os dados de endereço
# juntar os dataframes

del df['endereco2']
df = df.join(novo)
```

```python
# entender um puco mais o dataset

df.info()
```

```python
# Exibir os tipos de registros presentes no dataset
# Eu escolhi exibir dessa maneira pois o dataset é pequeno

print('data_consulta:',str(df['data_consulta'].drop_duplicates().values).strip(),'\n')
print('tipo_imovel:',str(df['tipo_imovel'].drop_duplicates().values).strip(),'\n')
print('area:',str(df['area'].drop_duplicates().values).strip(),'\n')
print('quarto:',str(df['quarto'].drop_duplicates().values).strip(),'\n')
print('banheiro:',str(df['banheiro'].drop_duplicates().values).strip(),'\n')
print('vagas_garagem:',str(df['vagas_garagem'].drop_duplicates().values).strip(),'\n')
print('uf:',str(df['uf'].drop_duplicates().values).strip(),'\n')
print('cidade:',str(df['cidade'].drop_duplicates().values).strip(),'\n')
print('bairro:',str(df['bairro'].drop_duplicates().values).strip(),'\n')
print('valor:',str(df['valor'].drop_duplicates().values).strip(),'\n')
```

```python
# Adicionar uma barra antes do valor do imovel para facilitar na divisão dos campos de aluguél e valor

df.loc[df['valor'].str.count('/') == 0, 'valor'] = '/' + df['valor'].astype(str)
```

```python
# separar os campos

novo_val = df['valor'].str.split("/", n = 2,  expand = True)
```

```python
# renomear as colunas e exibir a tabela

colunas = ['aluguel','valor_imovel']
novo_val.columns = colunas
novo_val.head(15)
```

```python
# formatar os dados de valor

def transform_str(value:str):
    return value.replace(',','').replace('R$','').replace('.',',').replace('Mês',',').replace('\n',',').replace(' ',',')

for col in novo_val.columns:
    novo_val['valor_imovel']=[transform_str(k) for k in novo_val['valor_imovel']]
    novo_val['aluguel']=[transform_str(k) for k in novo_val['aluguel']]
```

```python
# exibir os dados

novo_val.head(15)
```

```python
# exibir todos os valores

print('valor_imovel:',str(novo_val['valor_imovel'].drop_duplicates().values).strip(),'\n')
```

```python
# apagar a coluna de valor do imóvel do dataset original e juntar ao novo dataset que contem os valores limpos

del df['valor']
df = df.join(novo_val)
```

```python
# limpar os campos quarto e garagem

df['quarto'].replace('--','')
df['vagas_garagem'].replace('--','')
df.info()
```

```python
# converter em dados numericos

colunas = ['quarto','banheiro','vagas_garagem','area','aluguel','valor_imovel']
df[colunas] = df[colunas].apply(pd.to_numeric, errors='coerce')
```

```python
# exibir os tipos de dados do dataframe

df.info()
```

```python
df['valor_imovel'] = df['valor_imovel'].astype('float64')
```

```python
# Descrever o dataframe

df.describe()
```

```python
# exibir os promeiros registros do dataframe

df.head()
```

```python
# ajustar os outliers provocados por erros no registro

df.loc[df['area'] == 180000, 'area'] = 180
df.loc[df['area'] == 80000, 'area'] = 80
df.loc[df['area'] == 10, 'area'] = 100
```

```python
# resumo do dataframe

df.describe()
```

```python
# ajustar o campo bairro para tirar os espaços em branco

df['bairro'] = df['bairro'].str.strip()
```

### Dados limpos
Vamos gravá-los no banco novamente

```python
# gravar a saída na tabela do SQLite

df.to_sql('houseprices_vivareal_cleaned',
          conn,
          if_exists='replace',
          index=False)
```
